# Merging datasets 

```python
#To combine those DataSets into a single DataFrame, we can use the pd.merge() function:
#We started by combining transaction and cab datasets , pd.merge function recognizes that
#each DataFrame has an "transaction id" column, and automatically joins using this column as a key.
dataset1=pd.merge(transaction,cab)
dataset2=pd.merge(dataset1,city)
dataset2
#The result of the merge is a new DataFrame that combines the information from the two inputs.
```

```python
globaldataset=pd.merge(dataset2,customer)
globaldataset
#We merged the new DataFrame with the customer dataset in order to get a Master DataFrame that combines all the inputs .
```

```python
# Now we have to explore our new dataset :
globaldataset.shape
```

```python
for co in globaldataset.columns:
    if ' ' in co:
        globaldataset=globaldataset.rename(columns={co:co.replace(' ','_')})
        
#we replaced the space with an underscore 
```

#Cleaning & preparing Data

```python
minimum=globaldataset['Date_of_Travel'].min()
minimum
```

```python
maximum=globaldataset['Date_of_Travel'].max()
maximum
```

```python
difference=maximum-minimum
difference
```

```python
from datetime import datetime, timedelta , date
```

```python
globaldataset["Date_of_Travel"]=globaldataset.apply(lambda x:43465-x["Date_of_Travel"],axis=1)
globaldataset
```

```python
startDate=datetime(2016,1,31)
```

```python
globaldataset["Date_of_Travel"]= startDate+globaldataset["Date_of_Travel"].map(timedelta)
```

```python
globaldataset
```

```python
globaldataset["Date_of_Travel"].max()
```

```python
globaldataset["Date_of_Travel"].min()
```

```python
#we will create other culomns to separte months and years and ignore days and add them to our dataframe .
```

```python
#declaring two new variables
Months=[]
Years=[]
for k in range(len(globaldataset['Date_of_Travel'])):
    Months.append(globaldataset['Date_of_Travel'][k].month)
    Years.append(globaldataset['Date_of_Travel'][k].year)
globaldataset['Month']=Months
globaldataset['Year']=Years
```

```python
globaldataset.tail()
```

```python
# All we have to do now is dropping the date of travel's column
globaldataset.drop(["Date_of_Travel"], axis=1)
```

```python
globaldataset.groupby(["Year"]).mean()
```

```python
globaldataset.groupby(["Month"]).mean()
```

```python
#adding new column for the profit (difference between cost of the trip and the price charged)
globaldataset["Profit"]=(globaldataset["Price_Charged"]-globaldataset["Cost_of_Trip"])
globaldataset.tail()
```
